## Hands On 01


Often in the context of trees, a node with degree 1 is called a *leaf*. Write a function named `get_leaves` that takes a graph as an argument, loops through the nodes, and returns a list of nodes with degree 1.

```python
import networkx as nx

def get_leaves(G):
  # PUT YOUR CODE HERE
  # Primeiro declaramos a variável que conterá as folhas do grafo
  leaves = []
  # Percorremos os nós do grafo// node é o nó // val é o grau daquele nó
  for (node, val) in G.degree():
      # Verificamos se o grau do nó é 1 (ou seja, se ele é uma folha)
      if val == 1:
          # Adicionamos a variável que declaramos lá no inicio da função
          leaves.append(node)
    # Retornamos as folhas do grafo
  return leaves
```

```python
# this block of code is only for networkx adapt it for graph-tool
G = nx.Graph()
G.add_edges_from([
        ('a', 'b'),
        ('a', 'd'),
        ('c', 'd'),
    ])
assert set(get_leaves(G)) == {'c', 'b'}
```

## Hands On 02


Write a function max_degree that takes a graph as its argument, and returns a 2-tuple with the name and degree of the node with highest degree.

```python
import networkx as nx

def max_degree(G):
    # Declaramos a variável que conterá o nó de maior grau
    max = None
    # Percorremos os nós do grafo. node é o nó e val é o grau daquele nó
    for (node, val) in G.degree():
        # Verificamos se a variável max está vazia, pois se estiver qualquer valor é um nó de maior grau até agora
        if max is None:
            # Declaramos o nó de maior grau (inicial) e vamos para a próxima iteração
            max = (node, val)
            continue
        
        # Verificamos se o nosso nó é menor que o grau do nó que estamos percorrendo
        if max[1] < val:
            # Declaramos o novo nó de maior grau
            max = (node, val)
    # Retornamos o nó de maior grau que encontramos
    return max
```

```python
# this block of code is only for networkx adapt it for graph-tool
SG = nx.read_adjlist('friends.adjlist')
assert max_degree(SG) == ('Claire', 4)
```

## Hands on 03


Write a function `mutual_friends` that takes a graph and two nodes as arguments, and returns a list (or set) of nodes that are linked to both given nodes. For example, in the graph `SG` drawn above,

    mutual_friends(SG, 'Alice', 'Claire') == ['Frank']

an empty list or set should be returned in the case where two nodes have no mutual friends, e.g. George and Bob in `SG` drawn above.

```python
def mutual_friends(G, node_1, node_2):
    # Lista com os valores do nó 1
    list_1 = []
    # Percorremos os nós que se relacionam com o nó 1
    for node, val in SG.edges(node_1):
        # Adicionamos os valores desses nós a lista que declararmos anteriormente
        list_1.append(val)
    # Lista com os valores do nó 2
    list_2 = []
    # Percorremos os nós que se relacionam com o nó 2
    for node, val in SG.edges(node_2):
        # Adicionamos os valores desses nós a lista que declararmos anteriormente
        list_2.append(val)
    # Pegamos a intersecção das listas
    set_intersection = set(list_1).intersection(set(list_2))
    # Retornamos a lista com os nós que estão na intersecção
    return list(set_intersection)
```

```python
# this block of code is only for networkx adapt it for graph-tool
SG = nx.read_adjlist('friends.adjlist')
assert mutual_friends(SG, 'Alice', 'Claire') == ['Frank']
assert mutual_friends(SG, 'George', 'Bob') == []
assert sorted(mutual_friends(SG, 'Claire', 'George')) == ['Dennis', 'Frank']
```
